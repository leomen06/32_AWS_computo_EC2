```bash
#!/bin/bash

# Use this for your user data (script from top to bottom)

# install httpd (Linux 2 version)

yum update -y

yum install -y httpd

systemctl start httpd

systemctl enable httpd

echo "<h1>Hello World from $(hostname -f)</h1>" > /var/www/html/index.html
```

## Nota_01: 

**httpd is the Apache HyperText Transfer Protocol (HTTP) server program. It is designed to be run as a standalone daemon process. When used like this it will create a pool of child processes or threads to handle requests.

In general, httpd should not be invoked directly, but rather should be invoked via apachectl on Unix-based systems or as a service on Windows NT, 2000 and XP and as a console application on Windows 9x and ME.**

Fuente: https://httpd.apache.org/docs/2.4/programs/httpd.html

El script proporcionado es un script de usuario (user data) que se ejecutaría al lanzar una instancia en AWS EC2 con una AMI de Amazon Linux 2. Este script instala y configura el servidor web Apache (httpd) y crea una página web simple que muestra "Hello World" junto con el nombre del host completo de la instancia.

Aquí hay una explicación paso a paso del script:

1. **Comentario:**
   ```bash
   #!/bin/bash
   ```
   - Este es el shebang que indica que el script debe ejecutarse con Bash.

2. **Comentario:**
   ```bash
   # Use this for your user data (script from top to bottom)
   ```
   - Comentario que indica que este script puede utilizarse como datos de usuario al lanzar una instancia.

3. **Actualizar y instalar httpd:**
   ```bash
   yum update -y
   yum install -y httpd
   ```
   - `yum update -y`: Actualiza todos los paquetes instalados en el sistema.
   - `yum install -y httpd`: Instala el servidor web Apache (httpd).

4. **Iniciar y habilitar el servicio httpd:**
   ```bash
   systemctl start httpd
   systemctl enable httpd
   ```
   - `systemctl start httpd`: Inicia el servicio httpd.
   - `systemctl enable httpd`: Configura httpd para que se inicie automáticamente en el arranque del sistema.

5. **Crear la página HTML:**
   ```bash
   echo "<h1>Hello World from $(hostname -f)</h1>" > /var/www/html/index.html
   ```
   - `echo "<h1>Hello World from $(hostname -f)</h1>"`: Crea una cadena que contiene un encabezado HTML con el nombre completo del host de la instancia.
   - `> /var/www/html/index.html`: Redirige la cadena al archivo `index.html` en el directorio de documentos raíz del servidor web Apache.
   
En resumen, este script Bash se utiliza para configurar un servidor web Apache en una instancia de Amazon Linux 2 y crear una página web simple que muestra un saludo "Hello World" junto con el nombre del host completo de la instancia. Este script puede ser útil al lanzar instancias de manera automatizada en AWS EC2.

## Nota_02:

El parámetro `-f` no está directamente relacionado con el comando `echo`. En este contexto, `hostname -f` es un subcomando que se utiliza para obtener el nombre completo del host (fully qualified domain name, FQDN). Se puede checkear buscando `man hostname`

El subcomando `hostname -f` está incrustado en la cadena de texto dentro de la expansión de comillas dobles `$(...)`. Lo que hace es ejecutar el comando `hostname -f` para obtener el nombre completo del host y luego sustituir esa expresión con el resultado antes de que la cadena se pase como argumento al comando `echo`.

Entonces, en términos sencillos, `$(hostname -f)` se sustituirá con el FQDN del host actual cuando se ejecute el script.

El comando `echo` simplemente imprime la cadena resultante en el archivo `/var/www/html/index.html`. El operador `>` se utiliza para redirigir la salida del comando `echo` y escribirla en el archivo especificado, en este caso, `index.html` en el directorio `/var/www/html/`.

En resumen, `-f` en este contexto pertenece al comando `hostname` y significa "nombre completo del host". Recordar que `$(comando)` ejecuta el comando dentro de la cadena.

`*************************************************************************************************************************`

En el contexto de los grupos de seguridad en AWS, inbound y outbound se refieren al tráfico que entra y sale de un recurso protegido por un grupo de seguridad.

**Inbound** se refiere al tráfico que entra en un recurso protegido por un grupo de seguridad. Las reglas de entrada definen qué tráfico se permite que entre en un recurso.

**Outbound** se refiere al tráfico que sale de un recurso protegido por un grupo de seguridad. Las reglas de salida definen qué tráfico se permite que salga de un recurso.

Las reglas de entrada y salida se definen mediante el protocolo, el puerto y la dirección IP o el rango de direcciones IP. Por ejemplo, una regla de entrada podría permitir que el tráfico HTTP entrante desde cualquier dirección IP se conecte a un servidor web. Una regla de salida podría permitir que el tráfico TCP saliente desde un servidor web se conecte a cualquier dirección IP en el puerto 80.

Los grupos de seguridad se utilizan para controlar el acceso a los recursos protegidos en AWS. Las reglas de entrada y salida permiten que solo el tráfico autorizado acceda a los recursos.

El Protocolo de control de transmisión (TCP) es un estándar de comunicaciones que permite que los programas de aplicaciones y los dispositivos informáticos intercambien mensajes a través de una red. Está diseñado para enviar paquetes a través de Internet y garantizar la entrega exitosa de datos y mensajes a través de redes.

Aquí hay algunos ejemplos de cómo se pueden utilizar las reglas de entrada y salida:

* **Para permitir que los usuarios de la web accedan a una aplicación web:** Se podría crear una regla de entrada que permita el tráfico HTTP entrante desde cualquier dirección IP.
* **Para permitir que los clientes se conecten a una base de datos:** Se podría crear una regla de entrada que permita el tráfico TCP entrante desde direcciones IP específicas.
* **Para permitir que los servidores se comuniquen entre sí:** Se podría crear una regla de salida que permita el tráfico TCP saliente desde un servidor a otro servidor.

Es importante configurar las reglas de entrada y salida de los grupos de seguridad con cuidado para garantizar que solo el tráfico autorizado tenga acceso a los recursos.

_ 1000	Sets the sticky bit

_ 2000	Sets the setgid bit

_ 4000	Sets the setuid bit

En el contexto de los permisos de archivos en sistemas operativos basados en Unix/Linux, los valores numéricos 1000, 2000 y 4000 corresponden a configuraciones especiales conocidas como "sticky bit," "setgid bit" y "setuid bit," respectivamente. Estas configuraciones afectan la forma en que se ejecutan los programas o cómo se accede a los archivos en el sistema.

1. **Sticky Bit (1000):**
   - **Representación Numérica:** 1000
   - **Significado:** Cuando se establece el sticky bit en un directorio, solo el propietario del archivo puede eliminar o mover sus propios archivos dentro de ese directorio. Otros usuarios, aunque tengan permisos para escribir en el directorio, no pueden eliminar o mover los archivos de otros usuarios.

2. **Setgid Bit (2000):**
   - **Representación Numérica:** 2000
   - **Significado:** Cuando se establece el setgid bit en un directorio, los archivos creados en ese directorio heredan el grupo del directorio en lugar del grupo primario del usuario que los creó. Para archivos ejecutables, esto significa que se ejecutarán con los permisos del grupo del directorio en lugar del grupo del usuario.

3. **Setuid Bit (4000):**
   - **Representación Numérica:** 4000
   - **Significado:** Cuando se establece el setuid bit en un archivo ejecutable, el programa se ejecutará con los privilegios del propietario del archivo en lugar de los privilegios del usuario que lo está ejecutando. Esto es útil en situaciones donde se necesita que un programa tenga permisos elevados para realizar ciertas tareas, pero no se desea otorgar esos permisos al usuario que lo ejecuta.

Estos bits especiales permiten configuraciones avanzadas de permisos y son representados numéricamente por 1000, 2000 y 4000, respectivamente, en la notación octal utilizada por el comando `chmod` en sistemas Unix/Linux.

## Algo para tener en cuenta, Referenciar un grupo de seguridad desde otro grupo de seguridad:

No, no puedes llamar a un grupo de seguridad, esto no tiene ningún sentido.

Si un grupo de seguridad A hace referencia al grupo de seguridad B, significa que la instancia a la que está asociado el grupo de seguridad A permite el tráfico entrante o saliente a otra instancia a la que está asociado el grupo de seguridad B. Pero si desea enviar tráfico desde la instancia con el grupo de seguridad A a la instancia con el grupo de seguridad B, debe usar la IP o el DNS de la instancia B.

Generalmente, la razón por la que querríamos hacer referencia a un grupo de seguridad en lugar de una IP es que la IP puede cambiar con el tiempo o no está expuesta en absoluto. Un ejemplo bastante común es tener un balanceador de carga de aplicaciones (ALB) y un grupo de instancias EC2 que permiten el tráfico solo desde el ALB. La dirección IP del ALB cambia con el tiempo, por lo que para poder recibir tráfico del ALB, podemos hacer referencia al grupo de seguridad adjunto.

Si queremos hacer referencia a un grupo de seguridad de otro grupo de seguridad, tenemos que editar las reglas del grupo de seguridad inicial:

### <a name='mark_15'></a>Opciones de compra de instancias
### [Indice](#index)

![](img_90.png)

## Instancias bajo demanda: 

- Cargas de trabajo cortas y precios predecibles.
- Pago por uso (Linux o Windows facturación por segundos, también después del primer minuto que encendemos la máquina) 
- Costos más altos. 
- Flexibilidad para apagar en cualquier momento.
- Sin compromisos o contratos largos con AWS.
- Recomendado para cargas interrumpidas y de corto plazo, no se puede predecir como se comportará la aplicación.

## Instancias reservadas 

- Mínimo se pueden reservar por 1 año.
- Se puede reservar entre 1 a 3 años. 
- Descuentos de hasta 72% en comparación con las de "bajo demanda".
- Recomandadas para uso estable como DB. 
- tienen 2 modelos: 
    - Instancias Reservadas Convertibles: --> Se puede cambiar el tipo de instancia cuando lo necesitemos, hasta 45% descuento. 
    - Instancias Reservas Programadas: --> Se lanzan por un período programado (ejm: black friday)

## Instancias Spot - no recomendable para ambientes de producción

- Cargas de trabajo cortas y economicas (hay una especie de subastas de instancias, donde otros usuarios buscan instancias a menor precio)

## Host Dedicados 

- Servidor fisico, con capacidad de instancia EC2. 
- Se puede traer sus propias licencias (temas de compliance), instalarlo en el host. 
- se puede reservar hasta 3 años.

## Instancias dedicadas: 

- Se ejecutaran en un Hardware dedicado 
- Se puede compartir el Hardware con otras instancias dentro de la misma cuenta, sin control sobre la ubicación de las instancias, podemos mover el hardware después de tenerlo e iniciarlo.
